## Q3

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import cumtrapz
```

```python
earnings = [42000, 20000, 31000, 24000, 34000, 37000, 29000, 35000, 60000, 32000]
```

```python
len(earnings)
```

```python
# get quintiles and draw lorenz curve
earnings.sort()
quintiles = np.percentile(earnings, [20, 40, 60, 80])
quintiles
```

```python
cum_sum = np.cumsum(earnings) / sum(earnings)
# need to insert 0
cum_sum = np.insert(cum_sum, 0, 0)
```

```python
plt.plot(np.linspace(0, 1, len(cum_sum)), cum_sum, marker='o')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Wage quantile')
plt.ylabel('Cumulative wage share')
plt.title('Lorenz curve')
plt.savefig('lorenz.pdf')
```

```python
# find area under lorenz curve
area = cumtrapz(cum_sum, dx=1/(len(cum_sum)))
# get total area of triangle
total_area = 0.5
# get gini coefficient
gini = (total_area - area[-1]) / total_area
```

```python
A = 50*[10]
B_init = 25*[5]
C_init = 25*[5]

B_after = 20*[7]
C_after = 30*[4.5]

# plot initial and final distributions
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].hist([A, B_init, C_init], bins=5, label=['A', 'B', 'C'])
ax[0].set_title('Initial distribution')
ax[0].legend()
ax[1].hist([A, B_after, C_after], bins=5, label=['A', 'B', 'C'])
ax[1].set_title('Final distribution')
ax[1].legend()
# compute std of initial and final distributions
std_init = np.std(A + B_init + C_init)
std_after = np.std(A + B_after + C_after)
```

```python
# get 80-20 difference
init_80_20 = np.percentile(A + B_init + C_init, 80) - np.percentile(A + B_init + C_init, 20)
after_80_20 = np.percentile(A + B_after + C_after, 80) - np.percentile(A + B_after + C_after, 20)
init_80_20, after_80_20
```
